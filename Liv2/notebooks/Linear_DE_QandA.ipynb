# Q&A: What is a Linear Differential Equation?\n
\n
**Your Question:** *What's a linear DE (short for Differential Equation)?*\n
\n
Based on the snippet from the Liv2 notes you shared:\n
$$ f'(x) = \\lambda(x)f(x), \\quad f(0) = 1 $$\n
\n
### Theoretical Explanation\n
A **differential equation (DE)** is simply an equation that relates an unknown function (here, $f(x)$) to its rate of change (its derivative, $f'(x)$). \n
\n
It is called **linear** because the unknown function $f(x)$ and its derivative $f'(x)$ only appear to the **first power** (no $f(x)^2$, $\\sqrt{f(x)}$, or $f(x) \\cdot f'(x)$).\n
\n
In the context of **Liv2 (Actuarial Mathematics)**, this specific linear DE describes a quantity that changes at a rate proportional to its current size. \n
- $\\lambda(x)$ is the **variable coefficient**. In life insurance, this could be the **force of mortality ($\\mu_x$)** or a continuously varying **interest rate ($\\delta_x$)**.\n
- The solution, $f(x) = \\exp \\left( \\int_0^x \\lambda(u) du \\right)$, tells us the total accumulated growth or survival probability by multiplying (integrating) these rates over time.\n
\n
---\n
### Analogy time!\n
Think of your bank balance growing with compound interest. The rate at which your money grows ($f'(x)$) depends on the current interest rate ($\\lambda(x)$) multiplied by how much money you currently have ($f(x)$). Because it's a *variable* coefficient, imagine your bank changes the interest rate every single day (or continuously)!\n
\n
Or alternatively, think of a group of people surviving over time. As time passes, they "leave" the group at some continuous rate $\\lambda(x)$. The rate at which the surviving population shrinks depends on how many are currently alive ($f(x)$) times the current death rate ($\\lambda(x)$).

```python
import numpy as np\n
import matplotlib.pyplot as plt\n
\n
# Let's model this practically!\n
# We'll imagine lambda(x) is a linearly increasing force of mortality: lambda(x) = 0.05 + 0.01 * x\n
# The solution is f(x) = exp( integral_0^x -(0.05 + 0.01*u) du )\n
# Remember for survival probability we'd have a negative rate: f'(x) = -lambda(x) f(x) \n
# Let's use positive growth for this example (like compound interest)\n
\n
x_values = np.linspace(0, 10, 100)\n
\n
# The analytical solution: integral of 0.05 + 0.01*u is 0.05*x + 0.005*x^2\n
f_x_solution = np.exp(0.05 * x_values + 0.005 * x_values**2)\n
\n
plt.figure(figsize=(8, 5))\n
plt.plot(x_values, f_x_solution, label="f(x) - Accumulated Value", color='blue')\n
plt.title("Growth of f(x) with variable rate λ(x) = 0.05 + 0.01x")\n
plt.xlabel("Time (x)")\n
plt.ylabel("Value f(x)")\n
plt.legend()\n
plt.grid(True)\n
plt.show()\n
```
