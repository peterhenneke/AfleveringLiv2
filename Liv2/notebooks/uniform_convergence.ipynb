# Uniform Convergence

This notebook provides a visual and interactive understanding of **uniform convergence** compared to pointwise convergence.

## Theory: The $\epsilon$-tube
A sequence of functions $f_n(x)$ converges uniformly to a limit function $f(x)$ on an interval if, for *any* width $\epsilon > 0$, you can find an integer $N$ where the *entire* graph of $f_n(x)$ stays inside an $\epsilon$-tube around $f(x)$ for all $n \geq N$.

This means the sequence gets close to the limit at the **same speed** everywhere.

### Example 1: Uniform Convergence
Let $f_n(x) = \frac{1}{n} \sin(nx)$ on $[0, 2\pi]$.
The limit function is $f(x) = 0$.

Notice how increasing $n$ compresses the amplitude. Eventually, the entire blue wave fits cleanly inside the red $\epsilon$-zone, no matter where you look on the x-axis.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider

def plot_uniform_convergence(n=1, epsilon=0.5):
    x = np.linspace(0, 2*np.pi, 500)
    f_n = (1/n) * np.sin(n * x) if n > 0 else np.zeros_like(x)
    f_limit = np.zeros_like(x)
    
    plt.figure(figsize=(10, 5))
    plt.plot(x, f_limit, 'r--', label='Limit $f(x) = 0$')
    plt.fill_between(x, f_limit - epsilon, f_limit + epsilon, color='red', alpha=0.1, label=fr'$\pm \epsilon$ tube ($\epsilon={epsilon}$)')
    plt.plot(x, f_n, 'b', label=fr'$f_{{{n}}}(x) = \frac{{1}}{{{n}}}\sin({n}x)$')
    
    plt.ylim(-1.2, 1.2)
    plt.title('Uniform Convergence Example')
    plt.legend(loc='lower left')
    plt.grid(True)
    plt.show()

interact(plot_uniform_convergence, n=IntSlider(min=1, max=20, step=1, value=1), epsilon=FloatSlider(min=0.05, max=1.0, step=0.05, value=0.5));
```

### Example 2: Pointwise (Non-Uniform) Convergence
Let $f_n(x) = x^n$ on the interval $[0, 1]$.
- If $x < 1$, $x^n \to 0$.
- If $x = 1$, $1^n \to 1$.

The limit function $f(x)$ is $0$ everywhere except at $x=1$, where it "jumps" to $1$.
Because the function is continuous but pinned at $(1,1)$, a piece of the blue line will **always** drag outside the red $\epsilon$-tube near $x=1$ before the jump, no matter how big $n$ gets.

```python
def plot_pointwise_convergence(n=1, epsilon=0.2):
    x = np.linspace(0, 1, 500)
    f_n = x**n
    
    plt.figure(figsize=(10, 5))
    # Draw epsilon tube around the 0 part of the limit
    plt.fill_between(x, -epsilon, epsilon, color='red', alpha=0.1, label=fr'$\pm \epsilon$ tube around 0 ($\epsilon={epsilon}$)')
    
    plt.plot(x[:-1], np.zeros_like(x[:-1]), 'r--', label='Limit $f(x) = 0$ (for $x < 1$)')
    plt.plot(1, 1, 'ro', label='Limit $f(1) = 1$')
    plt.plot(x, f_n, 'b', label=fr'$f_{{{n}}}(x) = x^{{{n}}}$')
    
    plt.ylim(-0.2, 1.2)
    plt.title('Pointwise (Non-Uniform) Convergence Example')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.show()

interact(plot_pointwise_convergence, n=IntSlider(min=1, max=50, step=1, value=1), epsilon=FloatSlider(min=0.05, max=0.5, step=0.05, value=0.2));
```
